```python
Athletes_df = Athletes_df.drop(['Bib', 'Zip', 'Date', 'Age Group'
                                #, 'Percent Change', 'Max Temp', 'Min Temp', 'Mean Temp', 'Finish', 'Positive Split',
                                ], axis = 1)
```

Prediction

```python
from sklearn.model_selection import train_test_split
```

```python
Athletes_df.drop('Age',axis=1) #age is the dependant variable
```

```python
Athletes_df['Age'] #it is the outcome
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(Athletes_df.drop('Age',axis=1), 
                                                    Athletes_df['Age'], test_size=0.2, 
                                                    random_state=42)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
#cross val
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
mse = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv = 5)
```

```python
np.mean(mse)
```

```python
pred = regression.predict(X_test)
```

```python
sns.displot(pred-y_test, kde = True)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(pred, y_test)
```
